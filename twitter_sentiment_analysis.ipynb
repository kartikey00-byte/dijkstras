```python
import numpy as np
import pandas as pd
import re
import swifter
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
column_names = ['target', 'id', 'date', 'flag', 'user', 'text']
data = pd.read_csv('//content//dataset.csv', encoding = 'ISO-8859-1', names=column_names)
```

```python
data.isnull().sum()
```

```python
data['target'].value_counts()
```

```python
data.replace({"target": {4:1}}, inplace=True)
```

```python
data
```

```python
port_stem = PorterStemmer()
```

```python
def stemmning(content):
  stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)

  return stemmed_content
```

```python
data['peocessed_text'] = data['text'].swifter.apply(stemmning)
```

```python
data.head()
```

```python
data.rename(columns={'peocessed_text': 'processed_text'}, inplace=True)
```

```python
data.head()
```

```python
data.to_csv('processed_data.csv', index=False)
```

```python
new_data = pd.read_csv('processed_data.csv',encoding = 'ISO-8859-1')
```

```python
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
new_data.dropna(subset=['processed_text'], inplace=True)
```

```python
new_data.isnull().sum()
```

```python
X = new_data["processed_text"].values
Y = data['target'].values
```

```python
print(X.shape, Y.shape)
```

```python
Y = Y[:X.shape[0]]
```

```python
X = X[:Y.shape[0]]
```

```python
print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
import joblib
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_train)
acc = accuracy_score(Y_train, prediction)
print('Training accuracy: ', acc)
```

```python
test_prediction = model.predict(X_test)
test_acc = accuracy_score(Y_test, test_prediction)
print('Test accuracy: ', test_acc)
```
